```python
data=data.drop(['Healthcare_1'],axis=1)
```

```python
data.loc[data.Rooms>6]
```

```python
data.loc[data.Rooms==10,'Rooms']=2
data.loc[data.Rooms==19,'Rooms']=1
#data.loc[1454,'Rooms']=1
```

```python
data.loc[data.Rooms==0,'Rooms']=1
```

```python
def prepare_square(df):
    df.loc[df['Square']>(df['Rooms']*40+20),'Square']=df['Rooms']*40+20
    df.loc[df['Square']<(df['Rooms']*15),'Square']=df['Rooms']*15
    return df
```

```python
data=prepare_square(data)
```

```python
#for i in data.Rooms.unique():
#    data['LifeSquare']=data.loc[data['Rooms']==i].apply(lambda x:x['LifeSquare'].fillna(data.loc[data['Rooms']==i].median()))
#data.loc[(data.LifeSquare.isna())&(data['Rooms']==2)].apply(lambda x:x.fillna(data.loc[data['Rooms']==2].median()))
```

```python
def fillna_lifesquare(df,df_source):
    df['LifeSquare']=df['LifeSquare'].fillna(df_source['LifeSquare'].mean())# median()
    return df
```

```python
data=fillna_lifesquare(data,data)
```

```python
#data.groupby(['DistrictId','Rooms'])[['Price']].mean().reset_index()
stats1=data.groupby(['DistrictId','Rooms'],as_index=False)[['Price']].mean().rename(columns={'Price':'mean_price'})
```

```python
def join_stats1(df,stats1,df_source,mode='train'):    
    df=df.merge(stats1, on=['DistrictId','Rooms'],how='left')
    if mode=='test':
        df['mean_price']=df['mean_price'].fillna(df_source['mean_price'].mean())
    return df
```

```python
data=join_stats1(data,stats1,data)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
#features=data.loc[:,['Rooms','Square','LifeSquare','mean_price']]
features=data[['Rooms','Square','LifeSquare','mean_price']]
target=data['Price']
```

```python
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression as LR,Ridge
from sklearn.ensemble import RandomForestRegressor as RF
```

```python
model1=LR()
model2=Ridge()
model3=RF()
```

```python
%%time
model1.fit(X_train,y_train)
```

```python
%%time
model2.fit(X_train,y_train)
```

```python
%%time
model3.fit(X_train,y_train)
```

```python
y_pred1=model1.predict(X_test)
y_pred2=model2.predict(X_test)
y_pred3=model3.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error as MSE, r2_score as r2
```

```python
print("Линейная регрессия \n",MSE(y_test,y_pred1),r2(y_test,y_pred1))
```

```python
print("Ridge регрессия \n",MSE(y_test,y_pred2),r2(y_test,y_pred2))
```

```python
print("Случайный лес \n",MSE(y_test,y_pred3),r2(y_test,y_pred3))
```
