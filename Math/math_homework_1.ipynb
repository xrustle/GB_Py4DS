### Задание 2.1
Решите систему уравнений:

$\left\{\begin{array}{l}{3 x-2 y+5 z=7} \\ {7 x+4 y-8 z=3} \\ {5 x-3 y-4 z=-12}\end{array}\right.$

Линейная или нелинейная это система?<br> 
А каждое уравнение по отдельности?

#### Решение:

$$\Delta=\left|\begin{array}{ccc}{3} & {-2} & {5} \\ {7} & {4} & {-8} \\ {5} & {-3} & {-4}\end{array}\right|=3\left|\begin{array}{cc}{4} & {-8} \\ {-3} & {-4}\end{array}\right|+2\left|\begin{array}{cc}{7} & {-8} \\ {5} & {-4}\end{array}\right|+5\left|\begin{array}{cc}{7} & {4} \\ {5} & {-3}\end{array}\right|=3(-16-24)+2(-28+40)+5(-21-20)=$$

$$=-3\cdot40+2\cdot12-5\cdot41=-120+24-205=-301$$

Определитель не равен 0, следовательно система линейных уравнений имеет одно решение.

1) Найдем $x$ методом Крамера:

$$\Delta_{x}=\left|\begin{array}{ccc}{7} & {-2} & {5} \\ {3} & {4} & {-8} \\ {-12} & {-3} & {-4}\end{array}\right|=7\left|\begin{array}{cc}{4} & {-8} \\ {-3} & {-4}\end{array}\right|+2\left|\begin{array}{cc}{3} & {-8} \\ {-12} & {-4}\end{array}\right|+5\left|\begin{array}{cc}{3} & {4} \\ {-12} & {-3}\end{array}\right|=7(-16-24)+2(-12-96)+5(-9+48)=$$

$$=-7\cdot40-2\cdot108+5\cdot39=-280-216+195=-301$$

$$x=\frac{\Delta_{x}}{\Delta}=\frac{-301}{-301}=1$$

2) Найдем $y$. Подставим $x$ в два уравнения:

$$\left\{\begin{array}{l}{7+4 y-8 z=3} \\ {5-3 y-4 z=-12}\end{array} \Rightarrow\left\{\begin{array}{l}{4 y-8 z=-4} \\ {3 y+4 z=17}\end{array}\right.\right.$$

$$y=\frac{\Delta y}{\Delta}=\frac{\left|\begin{array}{cc}{-4} & {-8} \\ {17} & {4}\end{array}\right|}{\left|\begin{array}{cc}{4} & {-8} \\ {3} & {4}\end{array}\right|}=\frac{-16+136}{16+24}=\frac{120}{40}=3$$

3) Найдем $z$, подставив $y$ в уравнение $3y + 4z = 17$

$$9+4z=17$<br>$z=2$$

#### Ответ:
$x=1, y=3, z=2$. Система и каждое уравнение по отдельности являются линейными.

### Задание 2.2
Решите систему уравнений:

$\left\{\begin{array}{l}{x^{2}+y\cdot x-9=0} \\ {x-y / 5=0}\end{array}\right.$

Линейная или нелинейная это система?<br>
А каждое уравнение по отдельности?

#### Решение:

$$\left\{\begin{array}{l}{x^{2}+y \cdot x-9=0} \\ {x-\frac{y}{5}=0}\end{array} \Rightarrow\left\{\begin{array}{l}{x=\frac{1}{5} y} \\ {\frac{1}{25} y^{2}+\frac{1}{5} y^{2}-9=0}\end{array} \Rightarrow\left\{\begin{array}{l}{x=\frac{1}{5} y} \\ {\frac{6}{25} y^{2}=9}\end{array}\right.\right.\right.\Rightarrow\left\{\begin{array}{l}{y_{1,2}=\pm \frac{5 \sqrt{6}}{2}} \\ {x_{1,2}=\pm \frac{\sqrt{6}}{2}}\end{array}\right.$$

#### Ответ:
$x=\frac{\sqrt{6}}{2}, y=\frac{5 \sqrt{6}}{2}$ или $x=-\frac{\sqrt{6}}{2}, y=-\frac{5 \sqrt{6}}{2}$. Система уравнений не является линейной. Линейным является только второе уравенение.

### Задание 3
Площадь пола прямоугольной комнаты равна $48 м^2$, а его периметр равен $28 м$.<br> Найдите длину и ширину комнаты.

#### Решение:

Составим систему уравнений на основе условия задачи, где $a$ - длина, $b$ - ширина:

$$\left\{\begin{array}{l}{a \cdot b=48} \\ {2\cdot(a+b)=28} \\ {a\geqslant b>0}\end{array} \Rightarrow\left\{\begin{array}{l}{a=14-b} \\ {(14-b) \cdot b=48} \\ {a\geqslant b>0}\end{array} \Rightarrow\left\{\begin{array}{l}{a=14-b} \\ {b^{2}-14 b+48=0} \\ {a\geqslant b>0}\end{array}\right.\right.\right.$$

Решим квадратное уравнение, используя теорему Виета:

$$\left\{\begin{array}{l}{b_{1}+b_{2}=14} \\ {b_{1} b_{2}=48}\end{array} \Rightarrow\left\{\begin{array}{l}{b_{1}=6} \\ {b_{2}=8}\end{array}\right.\right.$$

Найдем $a$ из уравнения $a=14-b$:

$$\left\{\begin{array}{l}{a_{1}=8} \\ {a_{2}=6}\end{array}\right.$$

Мы приняли, что $a\geqslant b$, поэтому одно решение $(8,6)$

#### Ответ:
длина $8м$, ширина $6м$.

### Задание 4
Постройте на одном графике две кривые $y(x)$ для функции двух переменной $y(k,x)=cos(kx)$,<br> взяв для одной кривой значение $k=1$,<br> а для другой – любое другое $k$, не равное $1$.

#### Решение:

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
x = np.linspace(0, 10, 101)

y1 = np.cos(x)
y2 = np.cos(x/2)

plt.plot(x, y1)
plt.plot(x, y2)

plt.title("$y(k,x) = cos(kx)$")
plt.xlabel("x")
plt.ylabel("y")
plt.legend(labels=["$k = 1$", "$k = 0.5$"])

plt.show()
```
