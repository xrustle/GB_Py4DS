### Задание 1
Даны два вектора в трехмерном пространстве: $(10,10,10)$ и $(0,0,-10)$.<br>
* Найдите их сумму.<br>
* Напишите код на `Python`, реализующий расчет длины вектора, заданного его координатами.

#### Решение:
$\vec{a}=(10,10,10)$<br>
$\vec{b}=(0,0,-10)$<br>
$\vec{a}+\vec{b}=(10+0,10+0,10-10)=(10,10,0)$
#### Ответ:
$(10,10,0)$

```python
import numpy as np
def vec_len(x1, x2, x3):
    return np.sqrt(x1 * x1 + x2 * x2 + x3 * x3)
```

```python
a = (4, 2, -4)
print(f'Длина вектора a = {vec_len(*a)}')
```

### Задание 2
Почему прямые не кажутся перпендикулярными? (см.ролик)
#### Ответ:
Из-за разного масштаба по осям `x` и `y`.<br>Можно добавить `plt.axis('equal')` и прямые будут перпендикулярны.

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 21)
y = 3 * x + 1
y2= (-1 / 3) * x + 1
plt.plot(x, y)
plt.plot(x, y2)
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal');
```

### Задание 3
Напишите код на Python, реализующий построение графиков:
* окружности,
* эллипса,
* гиперболы.

```python
plt.style.use('fivethirtyeight')
```

```python
x0, y0 = 2, 2  # координаты центра окружности
R = 6  # радиус окружности
x = np.linspace(x0 - R, x0 + R, 1001)
y1 = np.sqrt(R**2 - (x - x0)**2) + y0
y2 = - np.sqrt(R**2 - (x - x0)**2) + y0
plt.plot(x, y1, 'm-')
plt.plot(x, y2, 'm-')

x0, y0 = 1, 1  # координаты центра эллипса
a, b = 9, 4  # длины полуосей эллипса
x = np.linspace(x0 - a, x0 + a, 1001)
y1 = np.sqrt((1 - (x - x0)**2 / a**2) * b**2) + y0
y2 = - np.sqrt((1 - (x - x0)**2 / a**2) * b**2) + y0
plt.plot(x, y1, 'b-')
plt.plot(x, y2, 'b-')

x0, y0 = 1, 2  # координаты центра гиперболы
a, b = 4, 2
x1 = np.linspace(-10, x0 - a, 1001)
y1 = np.sqrt(((x1 - x0)**2 / a**2 - 1) * b**2) + y0
y2 = - np.sqrt(((x1 - x0)**2 / a**2 - 1) * b**2) + y0
x2 = np.linspace(x0 + a, 10,  1001)
y3 = np.sqrt(((x2 - x0)**2 / a**2 - 1) * b**2) + y0
y4 = - np.sqrt(((x2 - x0)**2 / a**2 - 1) * b**2) + y0
plt.plot(x1, y1, 'r-')
plt.plot(x1, y2, 'r-')
plt.plot(x2, y3, 'r-')
plt.plot(x2, y4, 'r-')

plt.axis('equal')
plt.xlabel('x')
plt.xlabel('y');
```
