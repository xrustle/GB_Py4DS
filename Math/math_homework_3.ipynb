Даны два вектора в трехмерном пространстве: $(10,10,10)$ и $(0,0,-10)$.<br>
* Найдите их сумму.<br>
* Напишите код на `Python`, реализующий расчет длины вектора, заданного его координатами.

#### Решение:
$\vec{a}=(10,10,10)$<br>
$\vec{b}=(0,0,-10)$<br>
$\vec{a}+\vec{b}=(10+0,10+0,10-10)=(10,10,0)$
#### Ответ:
$(10,10,0)$

```python
import numpy as np
def vec_len(x1, x2, x3):
    return np.sqrt(x1 * x1 + x2 * x2 + x3 * x3)
```

### Задание 4.1
Решить систему: $\left\{\begin{array}{l}{y=x^{2}-1} \\ {e^{x}+x\cdot(1-y)=1}\end{array}\right.$

#### Решение
$e^{x}+x\cdot(2-x^2)-1=0$

Построим график для визуальной оценки

```python
def f(x):
    return np.exp(x) + x * (2 - x**2) - 1

def y(x):
    return x**2 - 1
```

```python
x = np.linspace(-3, 5, 100)

plt.plot(x, f(x))

plt.hlines(0, -3, 5, color='red')

plt.xlabel('x')
plt.ylabel('y')
plt.legend(labels=["f(x) = exp(x) + 2x - x^3 - 1"])
plt.grid()
```

```python
from scipy.optimize import fsolve
```

```python
x_roots = fsolve(f, [-2, 0, 2, 4])
```

```python
for i, x_root in enumerate(x_roots):
    print(f'{x_root, y(x_root)}')
```
